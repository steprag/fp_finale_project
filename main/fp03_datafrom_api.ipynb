```python
import pandas as pd
import requests

# Empty list to store combined data
combined_data5 = []
# Iterate through pages
api_url = f'https://unstats.un.org/sdgapi/v1/sdg/Target/Data?target=5.5&areaCode=250&page=1&pageSize=1000'
# Fetch JSON data for each page
response = requests.get(api_url)
api_info = response.json()
# Append the "data" array from each page to the combined list
combined_data5.extend(api_info['data'])
# Add the combined "data" to the original JSON object
api_sum = combined_data5
#
df_api = pd.DataFrame(api_sum)
```

```python
df_api.to_csv('dataset/raw_data/api_data5.5_raw.csv',index=False)
```

```python
# SÃ©lectionner les lignes oÃ¹ la colonne "series" est Ã©gale Ã  "IC_GEN_MGTL_19ICLS" ou "IC_GEN_MGTN_19ICLS"
df_api_filtered = df_api[df_api['series'].isin(["IC_GEN_MGTL_19ICLS", "IC_GEN_MGTN_19ICLS"])].copy()

df_api_filtered
```

```python
# CrÃ©er un DataFrame avec les colonnes spÃ©cifiÃ©es
df_api_filtered = df_api_filtered.loc[:, ['seriesDescription', 'timePeriodStart', 'value']].copy()

# Afficher le DataFrame filtrÃ©
df_api_filtered
```

```python
df_api_filtered.columns
```

```python
# Renommer la colonne 'timePeriodStart' en 'time_period'
df_api_filtered.rename(columns={'timePeriodStart': 'time_period'}, inplace=True)

# Convertir les valeurs de la colonne 'time_period' en entiers
df_api_filtered['time_period'] = df_api_filtered['time_period'].astype(int)

df_api_filtered
```

```python
df_api_filtered.rename(columns={'seriesDescription': 'serie_description'}, inplace=True)
```

```python
df_api_pivot = df_api_filtered.pivot(index='time_period', columns='serie_description', values='value')
df_api_pivot
```

```python
df_api_pivot.columns
```

```python
df_api_pivot.rename(columns={
    'Proportion of women in managerial positions - 19th ICLS (%)': '%_women_in_managerial_positions',
    'Proportion of women in senior and middle management positions - 19th ICLS (%)': '%_women_in_senior_middle_management'
}, inplace=True)
```

```python
df_api_pivot.dropna(inplace=True)
```

```python
df_api_pivot
```

```python
# Convertir les colonnes en types numÃ©riques
df_api_pivot['%_women_in_managerial_positions'] = pd.to_numeric(df_api_pivot['%_women_in_managerial_positions'], errors='coerce')
df_api_pivot['%_women_in_senior_middle_management'] = pd.to_numeric(df_api_pivot['%_women_in_senior_middle_management'], errors='coerce')

# Calculer la diffÃ©rence entre les deux colonnes et l'assigner Ã  une nouvelle colonne
df_api_pivot['%_women_in_junior_management'] = df_api_pivot['%_women_in_managerial_positions'] - df_api_pivot['%_women_in_senior_middle_management']

df_api_pivot
```

```python
df_api_pivot.shape
```

```python
df_api_pivot.dtypes
```

```python

df_api_pivot['%_women_in_managerial_positions'] = df_api_pivot['%_women_in_managerial_positions'].astype(float)
df_api_pivot['%_women_in_senior_middle_management'] = df_api_pivot['%_women_in_senior_middle_management'].astype(float)


df_api_pivot['%_women_in_managerial_positions'] = df_api_pivot['%_women_in_managerial_positions'].round(2)
df_api_pivot['%_women_in_senior_middle_management'] = df_api_pivot['%_women_in_senior_middle_management'].round(2)
```
