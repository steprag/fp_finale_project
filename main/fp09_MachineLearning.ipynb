# Model Selection Rationale

**Objective**:
The primary objective of our analysis is to predict the pay_gap within various communes, which represents the difference in mean salaries between genders. Given that pay_gap is a continuous numerical variable, our analysis falls into the realm of regression.

**Definition of Target Variable**:
The pay_gap is defined as the difference in mean salaries between male and female employees. It is expressed as a numerical value, which quantifies the extent of salary disparity. In some contexts, this gap is also represented as a percentage; however, for the purposes of our predictive modeling, we treat it as a direct numerical difference, making our prediction task a regression problem.

**Choice of Regression Model**:
Regression models are ideally suited for predicting numerical outcomes where the relationship between the input variables (features) and the output variable (target) is expected to be linear or can be modeled as a combination of linear associations. In our case, the regression model will help us understand how various salary components (such as mean salaries across different job categories and age groups) contribute to the gender pay gap.

**Model Selection**:
For our initial analysis, we employed a RandomForestRegressor, a type of ensemble learning technique that uses multiple decision trees to improve predictive accuracy and control over-fitting. RandomForest is particularly robust to outliers and does not assume a linear relationship between the features and the target, which can provide a more nuanced understanding of the complex factors influencing the pay gap.

**Other Models - Tree-Based Models**:
- **Decision Tree Regressor**: A non-linear model that divides the feature space into regions. For predictions, it uses the average target value in each region. It’s very intuitive but can easily overfit the data.
- **Random Forest Regressor**: An ensemble of decision trees, typically trained with the "bagging" method to improve stability and accuracy over a single decision tree. It reduces overfitting and is often more accurate.
- **Gradient Boosting Machines (GBM)**: Sequentially builds new models to correct errors made by previous models. GBM can be used for both regression and classification problems.
- **XGBoost, LightGBM, CatBoost**: Optimized gradient boosting libraries that handle large data sets more efficiently and are often more performant than standard GBM.

```python
# because our target "pay_gap" is continuous, so our model is Regression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error

from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import BaggingRegressor, RandomForestRegressor,AdaBoostRegressor, GradientBoostingRegressor
```

## Model training

```python
# train test split
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size = 0.20, random_state=0)
```

```python
# normalization
normalizer = MinMaxScaler()

normalizer.fit(X_train)
```

```python
X_train_norm = normalizer.transform(X_train)

X_test_norm = normalizer.transform(X_test)
```

```python
X_train_norm = pd.DataFrame(X_train_norm, columns = X_train.columns)
X_train_norm.head()
```

```python
X_test_norm = pd.DataFrame(X_test_norm, columns = X_test.columns)
X_test_norm.head()
```

## Random Forest

```python
forest = RandomForestRegressor(n_estimators=155,
                            max_depth=20,random_state=20)
```

```python
forest.fit(X_train_norm, y_train)
```

```python
forest2 = RandomForestRegressor(n_estimators=155,
                            max_depth=30,random_state=20)
forest2.fit(X_train_norm, y_train)
```

## Bagging and Pasting

```python
bagging_reg = BaggingRegressor(DecisionTreeRegressor(max_depth=20),
                            n_estimators=155,
                            max_samples = 1000,
                            random_state=20)
# without pasting
```

```python
bagging_reg.fit(X_train_norm, y_train)
```

```python
bagging_reg2 = BaggingRegressor(DecisionTreeRegressor(max_depth=20),
                            n_estimators=155,
                            max_samples = 1000,
                            bootstrap=False,
                            random_state=20)
# with pasting
bagging_reg2.fit(X_train_norm, y_train)
```

## Boosting

```python
ada_reg = AdaBoostRegressor(DecisionTreeRegressor(max_depth=20),
                            n_estimators=100)
```

```python
ada_reg.fit(X_train_norm, y_train)
```

## Model evaluation

metrics -- r2 score, RMSE
