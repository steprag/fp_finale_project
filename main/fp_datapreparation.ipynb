a.4 - Cleaning 'population' table

```python
population.head()
```

```python
population= population.rename(columns = {"CS1_8": "job_cat",
                                    "SEXE": "gender",
                                    "AGEQ65": "age",
                                    "NB": "total_population",
                            
                                    })

```

```python
correspondance_cat = {
    1: 'farmer',
    2: 'artisan_merchant_entrepreneur',
    3: 'executive',
    4: 'middle_management',
    5: 'employee',
    6: 'worker',
    7: 'retired',
    8: 'unemployed',
}

# Remplacer les valeurs numériques dans la colonne 'job_cat' par les catégories correspondantes
population['job_cat'] = population['job_cat'].replace(correspondance_cat)

# Arrondir les valeurs de la colonne 'population' à des chiffres entiers
population['total_population'] = population['total_population'].round().astype(int)
```

```python
population.head()
```

```python
population["age"].unique()
```

```python
population.columns
```

```python
# Définition de la fonction pour catégoriser les âges
def categoriser_age(age):
    if age >= 15 and age <= 24:
        return 'young'
    elif age >= 25 and age <= 49:
        return 'medium'
    else:
        return 'old'

# Appliquer la fonction à la colonne 'age' pour créer la nouvelle colonne 'age_cat'
population['age_cat'] = population['age'].apply(categoriser_age)

population.head()
```

```python
population_final = population.copy()
# Enregistrer le DataFrame dans un fichier CSV avec un chemin spécifique
population_final.to_csv('dataset/clean_data/population_final.csv', index=False)
```

```python
population.isna().sum()
```

```python
# Faire la somme de la colonne 'NB'
somme_nb = population['total_population'].sum()

print("Somme de la colonne 'total_population':", somme_nb)
```

```python
# Grouper par les colonnes spécifiées et sommer les valeurs de 'total_population'
population_clean = population.groupby(['CODGEO', 'LIBGEO', 'job_cat', 'gender', 'age_cat'], as_index=False)['total_population'].sum()

# Afficher les premières lignes du DataFrame résultant
population_clean.head()
```

```python
# Faire la somme de la colonne 'NB'
somme_nb = population_clean['total_population'].sum()

print("Somme de la colonne 'total_population':", somme_nb)
```

```python
# Enregistrer le DataFrame dans un fichier CSV avec un chemin spécifique
population_clean.to_csv('dataset/clean_data/population_clean.csv', index=False)
```
